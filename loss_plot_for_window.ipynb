```python
import matplotlib.pyplot as plt
import pandas as pd
import os

# Specify the path of the input file
input_file_path = '/home/raj/Rajarshi/Term Project/rajarshi_code/itransformer_file/epoch_loss_log_window_itransformer_model.txt'

# Load the data from the file
loss_data = pd.read_csv(input_file_path)

# Print the DataFrame to check if it's loaded correctly
print(loss_data)

# Specify the folder where you want to save the plots
output_folder = '/home/raj/Rajarshi/Term Project/notebook_files/saved_plots/loss_plots'

# Ensure the folder exists; if not, create it
os.makedirs(output_folder, exist_ok=True)

# Extract the base filename from the input file path without the extension
base_filename = os.path.splitext(os.path.basename(input_file_path))[0]

# Create a figure and axis for plotting
plt.figure(figsize=(10, 6))

# Plot the loss values for each window
for window in loss_data['Window'].unique():
    window_data = loss_data[loss_data['Window'] == window]
    plt.plot(window_data['Epoch'], window_data['Train_Loss'], marker='o', linestyle='-', label=f'Window {window}')

# Add titles and labels
plt.title('Training Loss Over Epochs for Each Window')
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.legend(ncol=2, loc='upper right', fontsize='small')  # Adjust location and fontsize as needed
plt.grid(True)

# Save the plot using the extracted base filename
filename = f'{base_filename}.png'
plt.savefig(os.path.join(output_folder, filename), format='png', dpi=300)

# Show the plot
plt.show()
```
